```python
# Crear matrices
A1 = np.array ([[1,0],[0,3]])
A2 = np.array ([[-9,0],[0,-9]])
A3 = np.array ([[1,5],[5,1]])
A4 = np.array ([[5,1],[1,5]])

x1 = np.linspace(-1,1,100)
x2 = np.linspace(-1,1,100)
X1,Y1 = np.meshgrid(x1,x2)
np.shape(X1)
Z = np.matmul(np.matmul(X1[0:2,:].T,A1),X1[0:2,:])
```

```python
ax=plt.contour(X1,Y1,Z)
plt.colorbar(ax,shrink=0.5,aspect=5)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
Z1 = Z.reshape(X1.shape)
surf= ax.plot_surface(X1, Y1, Z1)
fig.colorbar(surf,shrink=0.5,aspect=5)
plt.show()
```
