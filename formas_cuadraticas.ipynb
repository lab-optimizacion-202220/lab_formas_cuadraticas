# a)

Obtenga los vectores propios normalizados v1, v2 de la matriz, i.e, |vi| = 1. Grafique en el plano los
vectores λ1v1 y λ2v2, siendo λi el valor propio asociado al vector vi.


```python
def vector_valor (vpA,vcpA):
    #multiplicar vector propio 1 con valor propio1
    vect1 = vpA*vcpA[0]
    #multiplicar vector propio 2 con valor propio2
    vect2 = vpA*vcpA[1]
    ax = plt.axes()
    #dibujar los vectores 
    ax.arrow(0,0, vect1[0],vect1[1],color="r", head_width=0.5, head_length=0.5,label = "$ \lambda_1*v_1$")
    ax.arrow(0,0, vect2[0],vect2[1], head_width=0.5, head_length=0.5,label = "$ \lambda_2* v_2$")
    #fijar límites del plano
    plt.xlim(-10,10)
    plt.ylim(-10,10)
    #ubicacion del legend
    plt.legend(loc="best")
    #grilla
    plt.grid()
    #plotear
    plt.show()
```

```python

plt.title("Valores propios*Vectores propios matriz A1")
vector_valor(vpA,vcpA)
```

```python

plt.title("Valores propios*Vectores propios matriz A2")

vector_valor(vpA2,vcpA2)
```

```python

plt.title("Valores propios*Vectores propios matriz A3")
vector_valor(vpA3,vcpA3)
```

```python

plt.title("Valores propios*Vectores propios matriz A4")

vector_valor(vpA4,vcpA4)
```

# b)
Grafique dos curvas de contorno de tal manera que estas est´en sobre los vectores

```python
def graficas (A):

    #valor y vectores propios
    vpA,vcpA = linalg.eig(A)
    #ejes
    x1 = np.linspace(-10,10,100)
    x2 = np.linspace(-10,10,100)
    X1,Y1 = np.meshgrid(x1,x2)

    #  f(x) = x^T A x
    Z = np.matmul(np.matmul(X1[0:2,:].T,A),X1[0:2,:])
    #Contorno de la función
    ax=plt.contour(X1,Y1,Z)
    #multiplicar vector propio 1 con valor propio1
    vect1 = vpA*vcpA[0]
    #multiplicar vector propio 2 con valor propio2
    vect2 = vpA*vcpA[1]
    ax2 = plt.axes()
    #dibujar los vectores 
    ax2.arrow(0,0, vect1[0],vect1[1],color="r", head_width=0.5, head_length=0.5,label = "$ \lambda_1*v_1$", width=0.2)
    ax2.arrow(0,0, vect2[0],vect2[1], head_width=0.5, head_length=0.5,label = "$ \lambda_2* v_2$", width = 0.2)

    plt.legend(loc="best")
    plt.colorbar(ax,shrink=0.5,aspect=5)
    plt.grid()
    plt.show()
```

```python

plt.title("Matriz A1")

graficas(A1)
```

```python
plt.title("Matriz A2")

graficas(A2)
```

```python
plt.title("Matriz A3")

graficas(A3)
```

```python
plt.title("Matriz A4")
graficas(A4)
```

### Quinto Punto

Cree 2 funciones polinomiales de orden 2 qi(x, y), i = 1, 2 tal que qi(x, y) = 0 ∀(x, y) ∈ Bd D la frontera de la
 egión D, región definida como:

```python
def polinomios (A,Z1):
    plt.title("Contorno y valores propios*vectores propios")
    graficas(A)


    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    # A1
    Z51 = Z1.reshape(X1.shape)
    #plotear en 3D la superficie
    surf= ax.plot_surface(X1, Y1, Z51,cmap=cm.winter,
                        linewidth=0, antialiased=False)
    fig.colorbar(surf, shrink=0.5, aspect=5)
    
    plt.title("Superficie polinomio")
    plt.show()

    # plotear el contourf 
    ax2=plt.contourf(X1,Y1,Z1)
    plt.title("Contourf polinomio")
    plt.colorbar(ax2,shrink=0.5,aspect=5)

```
