```python
def vector_valor (vpA,vcpA):
    #multiplicar vector propio 1 con valor propio1
    vect1 = vpA*vcpA[0]
    #multiplicar vector propio 2 con valor propio2
    vect2 = vpA*vcpA[1]
    ax = plt.axes()
    #dibujar los vectores 
    ax.arrow(0,0, vect1[0],vect1[1],color="r", head_width=0.5, head_length=0.5,label = "$ \lambda_1*v_1$")
    ax.arrow(0,0, vect2[0],vect2[1], head_width=0.5, head_length=0.5,label = "$ \lambda_2* v_2$")
    #fijar límites del plano
    plt.xlim(-10,10)
    plt.ylim(-10,10)
    #ubicacion del legend
    plt.legend(loc="best",)
    #grilla
    plt.grid()
    #plotear
    plt.show()
```

```python
vector_valor(vpA,vcpA)
```

```python
vector_valor(vpA2,vcpA2)
```

```python
vector_valor(vpA3,vcpA3)
```

```python
vector_valor(vpA4,vcpA4)
```
